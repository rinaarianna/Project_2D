### Scientific Question: 
**Does dextromethorphan, a common over-the-counter cough syrup, target the same biological pathways in the brain to the known pathways ketamine in its use as an antidepressant?**

Ketamine, a well-known psychedelic pharmaceutical drug, functions as an NMDA Receptor antagonist. This blockage of the NMDA receptor is thought to be responsible for ketamine's antidepressant effects in patients sufferring from major depressive disorder (Hashimoto et al 2016).

The over-the-counter cough syrup known as dextromethorphan is also known to be an NMDA receptor antagonist (Cole et al 1989). This suggests that the  dextromethorphan may also be effective in the treatment of depression through similar biological pathways involving the NMDA receptor. The structures of ketamine and dextromethorphan can be found using PubChem (https://pubchem.ncbi.nlm.nih.gov/). According to the database website: 

"PubChem is the world's largest collection of freely accessible chemical information. Search chemicals by name, molecular formula, structure, and other identifiers. Find chemical and physical properties, biological activities, safety and toxicity information, patents, literature citations and more."


### Scientific Hypothesis:
**If dextromethorphan is known to behave similarly to ketamine in its activity as an NMDA receptor antagonist, then ketamine and dextromethorphan will have similar chemical structures and consequently the pharmaceuticals will equally dock with the NMDA receptor in a similar location when being used as an antidepressant.**

The two compound structures of ketamine and dextromethorphan were downloaded from PubChem and saved as a SDF file, which was used in both the fingerprinting as well as the table displaying the structures and data from PubChem of both compounds. 

The first cheminformatics method used in answering the scientific question and testing the hypothesis was Fingerprinting and Similarity Mapping. Fingerprinting is a cheminformatics tool that allows 2D and 3D molecular structures to be converted into a comparable numerical code. By comparing the codes between different chemicals, in this case dextromethorphan and ketamine, we are able to determine the level of similarity between them which is important in testing the scientific hypothesis proposed above.

The fingerprints were also presented in a similarity map. By displaying the fingerprints in a similarity map, we are able to visualize the similarity in structure between the two drugs. Using color, the map is able to highlight struvtural areas of similarity and differences between the two chemicals (Riniker et al 2013). 

The second anaysis performed was the bioinformatics tool of displaying the structures of both dextromethorphan and ketamine in a table, along with the other PubMed data on the two pharmaceuticals. This provides a stronger understanding of the compounds and their chemical structures, which is part of my scientific hypothesis.

## Part 2A: Reading in the Data Files and Performing a Cheminformatics Analysis 
### (Fingerprinting, Similarity Analysis, and Similarity Mapping)

**SDF Files**: The files containing the 3D structures of dextromethorphan and ketamine were downloaded from PubMed in the form of a SDF File. SDF files contain the complete description about the structural and molecular organization of molecules, and this information is stored in a structured format. The file describes each atom's location in relation to another, using x, y, and z coordinates to outline the locations of each molecule and thus store essential information about the molecular structure. The SDF file can then be used by Jupyter Notebook to read the structure of a molecule. 

**Molecular Fingerprints**: Molecular fingerprints encode the 3D structure of a compound or molecule into a binary digits, known as bits. This code is able to be used in analysis of the similarity between two molecules. More information on fingerprints can be found here: 
(https://openbabel.org/docs/dev/Fingerprints/intro.html) 

The first step in the code below was to upload and read in the molecules from an SDF File (ketamine and dextromethorphan). These molecules were then turned into molecular fingerprints. Using the fingerprints and the itertools package, we then found the fingerprint analysis between the two molecules and stored this in a matrix, as seen below. 

```python
# Here, we load in the SDF file made prior, 'drugs_SDF'.
# 'drugs_SDF' contains SDF format files for ketamine and dextromethorphan
m = Chem.SDMolSupplier('drugs_SDF')
SDFFile= 'drugs_SDF'

# Using the RDKit tool "PandasTools" to read in the SDF file and 
# return the data as a pandas dataframe.
drugs=PandasTools.LoadSDF(SDFFile)

# defining the length of the SDF file,
# or the number of compounds being ran in the file.
# in this case, we have two compounds- dextromethorphan and ketamine.
chemical_len = len(drugs)
# print SDF file to check that the file was imported properly into the notebook
print(drugs)
```

We then run a fingerprint similarity analysis to show the similarity scores between dexromethorphan and ketamine. We are using a topological fingerprint analysis because we are only comparing two structures.

```python
# Creating a Fingerprint for Dextromethorphan using SMILE format
mol= Chem.MolFromSmiles('COc1ccc2c(c1)C13CCCCC1C(C2)N(C)CC3')
# importing necessary tool from RDKit
from rdkit.Chem import rdmolops
# Creating the fingerprint in a bit string format
fp = rdmolops.RDKFingerprint(mol, fpSize=2048, minPath=1, maxPath=7).ToBitString()
print(fp)
```

```python
# Creating the fingerprint for Ketamine
refmol= Chem.MolFromSmiles('CNC1(c2ccccc2Cl)CCCCC1=O')
# importing necessary tools from RDKit 
from rdkit.Chem import rdmolops
# creating the fingerprint for the compound in a bit string
fp = rdmolops.RDKFingerprint(refmol, fpSize=2048, minPath=1, maxPath=7).ToBitString()
print(fp)
```

```python
# Uses bit averages to calculate the simililarity between structures

# upload smile format of the two compounds
ms = [Chem.MolFromSmiles('CNC1(c2ccccc2Cl)CCCCC1=O'), Chem.MolFromSmiles('COc1ccc2c(c1)C13CCCCC1C(C2)N(C)CC3')]

# call for the fingerprinting of each molecule from the smiles input above
fps = [Chem.RDKFingerprint(x) for x in ms]
print(fps)

# call for the similarity between each fingerprint 
DataStructs.FingerprintSimilarity(fps[0],fps[1])
```

In the code below, we first constructed a smiles format of the compounds from the sdf file for both ketamine and dextromethorphan. These formulas were then used in constructing a similarities map of the two compounds.

**Similarity Map Using Fingerprinting**: A similarity map for fingerprinting is used to visualize the atom similarity between molecules. This map uses the "weight" of each atom in a molecule to determine the color used in the map. A "weight" is calculated by creating a fingerprint of both molecules and removing the bits. If there is a decrease in structural similarity when the bits are removed, the structure is outlined in green. If there is an increase in the structural similarity when bits are removed, then the structure gets highlighted in pink. When there is no change in structural similarity during the removal of the bits in that region, the structure is highlighted in grey. We can use this to visualize the structural similiarities and differences between the two compounds. 

```python
# read the molecular formula (Smiles format) of the compounds 
m=Chem.MolFromMolFile('3D-ketamine.sdf')
Chem.MolToSmiles(m)
```

```python
## Similarity Map using fingerprinting

# import the smiles format of ketamine and dextromethorphan
# using ketamine as the reference molecule under the variable refmol
# and dextromethorphan for comparison as the variable mol  
mol= Chem.MolFromSmiles('COc1ccc2c(c1)C13CCCCC1C(C2)N(C)CC3')
refmol= Chem.MolFromSmiles('CNC1(c2ccccc2Cl)CCCCC1=O')

# import the tools necessary to draw a similarities map from rdkit
from rdkit.Chem.Draw import SimilarityMaps

# set parameters for map within the similarity map tool
# fpType- the parameter to call the desired types of atom pairs and torsions 
# fpType set to normal is calling the default 
# fpType set to bv is referencing the bit vectors in the fingerprint
# which will be used in determining the similarity shown in the highlighted colors
# pink (more similar), green (less similar), or grey (no change)
fp = SimilarityMaps.GetAPFingerprint(mol, fpType='normal')
fp = SimilarityMaps.GetTTFingerprint(mol, fpType='normal')
fp = SimilarityMaps.GetMorganFingerprint(mol, fpType='bv')
fig, maxweight = SimilarityMaps.GetSimilarityMapForFingerprint(refmol, mol, SimilarityMaps.GetMorganFingerprint)
```
